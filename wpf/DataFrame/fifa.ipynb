```
private static double Mean(double[] values)
{
  if (values is null)
    throw new ArgumentNullException(nameof(values));

  if (values.Length == 0)
    throw new ArgumentException($"{nameof(values)} must not be empty");

  double sum = 0;
  for (int i = 0; i < values.Length; i++)
    sum += values[i];

  double mean = sum / values.Length;

  return mean;
}
```

```
var pltAll = new ScottPlot.Plot();
var countries = new[] { "ARG", "BRA"};
foreach (var country in countries)
{
    var countryData = df.Rows
        .Select(row => new { Date = DateTime.Parse(row["Date"].ToString()), Rank = double.Parse(row[country].ToString()) })
        .ToList();

    // X축: 날짜, Y축: 국가 순위 데이터를 준비합니다.
    double[] dates = countryData.Select(d => d.Date.ToOADate()).ToArray();
    double[] ranks = countryData.Select(d => d.Rank).ToArray();

    // 국가 순위 데이터를 그래프로 그립니다.
    var sca = pltAll.Add.Scatter(dates, ranks);
    sca.LegendText = country;
}
pltAll.Title("FIFA Rankings");
pltAll.YLabel("Rank");

pltAll.Axes.DateTimeTicksBottom();
pltAll.XLabel("Date");

pltAll.ShowLegend();

pltAll
```

```
// calculate the regression line
ScottPlot.Plot myPlot = new();

var franceData = df.Rows
            .Where(row => DateTime.Parse(row["Date"].ToString()).Year >= 2001)
            .Select(row => new { Date = DateTime.Parse(row["Date"].ToString()), FRA = double.Parse(row["FRA"].ToString()) })
            .ToList();


// X축: 날짜, Y축: 프랑스 순위 데이터를 준비합니다.
double[] dates = franceData.Select(d => d.Date.ToOADate()).ToArray();
double[] franceRanks = franceData.Select(d => d.FRA).ToArray();

var sp = myPlot.Add.Scatter(dates, franceRanks);
sp.LineWidth = 0;
sp.MarkerSize = 10;

ScottPlot.Statistics.LinearRegression reg = new(dates, franceRanks);

// plot the regression line
Coordinates pt1 = new(dates.First(), reg.GetValue(dates.First()));
Coordinates pt2 = new(dates.Last(), reg.GetValue(dates.Last()));
var line = myPlot.Add.Line(pt1, pt2);
line.MarkerSize = 0;
line.LineWidth = 2;
line.LinePattern = LinePattern.Dashed;

// note the formula at the top of the plot
myPlot.Title(reg.FormulaWithRSquared);

myPlot.Axes.DateTimeTicksBottom();
myPlot
```
